```python
arianagrande="https://es.wikipedia.org/wiki/Ariana_Grande" 
IFrame(arianagrande, width=900, height=900)
```

```python
import pandas as pd

Arianaapd=pd.read_html(arianagrande, # link
                        header=0, # where is the header?
                        flavor='bs4')
```

```python
len(Arianaapd)
```

```python
Arianaa=Arianaapd[3].copy()
```

```python
Arianaa
```

```python
Arianaa=Arianaa.iloc[:, [0, 1, 2, 3]]
```

```python
Arianaa
```

```python
Arianaa=Arianaa.rename(columns={'Año': 'Year', 'Título': 'Title', 'Rol': 'Role', 'Notas': 'Notes'})
```

```python
Arianaa
```

```python
import os

Arianaa.to_csv(os.path.join("Tablas Limpias","Arianaa.csv"),index=False)
```
